```python
import pandas as pd

samples= pd.read_csv('seed_data1.csv') 
#labels=samples['varieties'].values
#labels
```

```python
# Perform the necessary
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt# Calculate the linkage: mergings
mergings = linkage(samples, method='complete')# Plot the dendrogram, using varieties as labels
dendrogram(mergings,
           leaf_rotation=90,
           leaf_font_size=6,)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
labels1= fcluster(mergings, 6, criterion='distance')
print(labels1)
```
